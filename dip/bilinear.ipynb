```python
import numpy as np
from scipy import interpolate

def bilinear_interpolation(img, new_shape):
    # Get the original image shape
    old_shape = img.shape

    # Create a 2D grid of coordinates for the old and new image
    x_old = np.linspace(0, old_shape[1]-1, old_shape[1])
    y_old = np.linspace(0, old_shape[0]-1, old_shape[0])
    x_new = np.linspace(0, old_shape[1]-1, new_shape[1])
    y_new = np.linspace(0, old_shape[0]-1, new_shape[0])

    # Create interpolation function for each color channel
    interp_func_r = interpolate.interp2d(x_old, y_old, img[:,:,0], kind='linear')
    interp_func_g = interpolate.interp2d(x_old, y_old, img[:,:,1], kind='linear')
    interp_func_b = interpolate.interp2d(x_old, y_old, img[:,:,2], kind='linear')

    # Evaluate interpolation function at new coordinates
    new_r = interp_func_r(x_new, y_new)
    new_g = interp_func_g(x_new, y_new)
    new_b = interp_func_b(x_new, y_new)

    # Stack color channels into new image
    new_img = np.dstack((new_r, new_g, new_b))

    # Return the new image
    return new_img



import cv2

# Load an image
img = cv2.imread('rgb.jpg')

# Resize the image using bilinear interpolation
new_shape = (img.shape[0]*4, img.shape[1]*2)
resized_img = bilinear_interpolation(img, new_shape)

# Display the original and resized images side by side
cv2.imshow('Original Image', img)
cv2.imshow('Resized Image', resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
