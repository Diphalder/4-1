```python
import os
import cv2

# Set the directory path containing JPEG images
input_dir = '/content'

# Create a directory for storing the output PNG images
output_dir = '/content'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through all the JPEG images in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.jpg') or filename.endswith('.jpeg'):
        # Load the image
        img_path = os.path.join(input_dir, filename)
        img = cv2.imread(img_path)

        # Convert the image to PNG format
        png_filename = os.path.splitext(filename)[0] + '.png'
        png_path = os.path.join(output_dir, png_filename)
        cv2.imwrite(png_path, img)
```

The os module is operating system functions for working with files and directories.

 os.listdir(input_dir) is used to travarse input_dir directory files and perform operations.



The os.path.join() function is then used to create the full file path of the output PNG file by joining the output_dir variable with the png_filename variable.

 cv2.imwrite() function is used to save the converted PNG image to the output directory using the png_path file path.
